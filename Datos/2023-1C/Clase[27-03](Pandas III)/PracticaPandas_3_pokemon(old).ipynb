Al cargar un dataset desde una hoja de calculos, si tiene mas de una hoja debo especificar que hoja voy a cargar, sino se carga la primera. 

Para cada hoja primero voy a obtener el nombre del pack que va a cargarse como el nombre de la primer columna; luego la fecha de lanzamiento se encuenta en la segunda columna de la primer fila, asi que lo puedo acceder con `iloc`. Una vez que obtuve esos datos, puedo recortar las filas extras haciendo un slice con `iloc` nuevamente. Para poder mantener control, a cada una le asigno un numero de carta dentro del set, y un numero de carta en la db general (el orden en que fueron agregadas)

https://docs.google.com/spreadsheets/d/10G8wEY70qJ7rEnGbDzJp13vdOB4ikIjz4E3SQVRglRc/edit#gid=967492039

```python
t_df = pd.read_excel("/content/Pokemon TCG Spreadsheet V3.23.xlsx",sheet_name=1,dtype=str)

t_df["pack"] = t_df.columns[0]
t_df["release_date"] = t_df.iloc[0][1].split()[0]
t_df = t_df.iloc[3:]
t_df.columns = ["set_id","name","type","a","b","c","d","e","pack","release_date"]
t_df = t_df[["set_id","name","type","pack","release_date"]]
t_df = t_df.dropna()
t_df.set_id = range(1,len(t_df)+1)
t_df["card_id"] = range(0+1,0+len(t_df)+1)

t_df
```

Siendo que son muchas hojas, pero todas las hojas son iguales, puedo automatizar lo anterior utilizando un `for` y acumulando todos los dataframes en una lista que luego puedo utilizar para concatenar

```python
dataframes = []
last_added = 0

for x in range(1,144):
    t_df = pd.read_excel("/content/Pokemon TCG Spreadsheet V3.23.xlsx",sheet_name=x,dtype=str)
    t_df["pack"] = t_df.columns[0]
    t_df["release_date"] = t_df.iloc[0][1].split()[0]
    t_df = t_df.iloc[3:]
    t_df.columns = ["set_id","name","type","a","b","c","d","e","pack","release_date"]
    t_df = t_df[["set_id","name","type","pack","release_date"]]
    t_df = t_df.dropna()
    t_df.set_id = range(1,len(t_df)+1)
    t_df["card_id"] = range(last_added+1,last_added+len(t_df)+1)
    last_added += len(t_df)
    dataframes.append(t_df)
    print("Added",len(t_df),"from",t_df.iloc[0].pack)
```

```python
database = pd.concat(dataframes)
database
```

Si reviso los tipos, me doy cuenta que hay errores de tipeo. Antes de poder cambiarle el tipo a la columna, deberia arreglar eso

```python
database.type.unique()
```

```python
def fix_types(x):
    if x.rstrip() in ["Colorelss","Colorless"]:
        return "Normal"
    if x in ["Lighting","Lightnijg","Electric"]:
        return "Lightning"
    if x in ["Fightning","FIGHting"]:
        return "Fighting"
    return x    

database.type = database.type.map(lambda x:fix_types(x))
database.type.unique()
```

Algo que ahora me serviria tener, es una columna que me identifique las cartas que son de pokemon y cuales no, ya que es una operacion que probablemente haga mucho y no es exactamente 'barata'

```python
not_pokemon_types = set(['Trainer', 'Energy', 'Stadium', 'Tool', 'Supporter', 'TM',  'Item'])

database["is_pokemon_card"] = database.type.map(lambda x:x not in not_pokemon_types)
database
```

Ahora si podemos combinar los dos dataframes para obtener un nuevo dataset mas completo. El problema es que no tengo como relacionar ambas tablas. Pero eso es algo que puedo construir trabajando los datos.

Yo se que el nombre del pokemon tiene que aparecer en algun lugar del nombre de la carta, entonces, si chequeo eso, puedo asociar a cada carta un nombre de pokemon. Mas aun, como se que el segundo dataframe contiene toda la info de todos los pokemon existentes, utilizo eso.

```python
pokenumber = dict(zip(small_dex.name.map(lambda x:x.lower()),small_dex.index))
```

```python
pokenumber
```

Me defino la funcion, y se la aplico a la columna name con `map` para obtener la nueva columna que me va a servir para hacer el `join`.

```python
def find_pokemon(x):
    x = x.lower()
    if x in pokenumber:
        return pokenumber[x]
    for y in pokenumber:
        if y in x:
            return pokenumber[y]
    return "ERROR"

database["join_id"] = database.name.map(find_pokemon)
database
```

La realidad es que si ya se que no es un pokemon, puedo mejorar el uso de recursos solo operando con las cartas que si son pokemon. Para eso, utilizo `loc`, tanto para asignar los valores como para acceder a los datos (ya que no genera una copia)

```python
database["join_id"] = np.nan
database.loc[database.is_pokemon_card,"join_id"] = database.loc[database.is_pokemon_card].name.map(find_pokemon)

database
```

Incluso ahora puedo discriminar las cartas que no deberian ser de pokemon de las que si pero tienen algun error de tipeo

Con ahora si una columna para poder unir, hago el join

```python
tcg_dataset = database.set_index("join_id").join(small_dex,how="left",rsuffix="_dex").reset_index()
tcg_dataset.set_index("card_id")
```

```python
tcg_dataset = tcg_dataset.rename(columns={"index":"pokemon_name"})
stats = tcg_dataset.loc[tcg_dataset.is_pokemon_card].groupby("pokemon_name").agg({"type":[lambda x: pd.Series.mode(x)[0],"nunique","count"],"release_date":{"max","min"}})
```
