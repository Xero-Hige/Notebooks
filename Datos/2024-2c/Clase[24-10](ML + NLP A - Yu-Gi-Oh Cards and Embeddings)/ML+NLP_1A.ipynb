```python
import re
regex = r"[^\w\d]+"

def clean_split(card_text):
    clean = re.sub(regex," ",card_text)
    return clean.lower().split()


ds["tokenized_text"] = ds.card_text.map(lambda x: clean_split(x))
ds.tokenized_text
```

```python
import random

for x in range(5):
    total_loss = 0

    for x in range(10,len(train),10):

        indexes = train[x-10:x]

        x = []
        y = []

        for index in indexes:
            row = ds.iloc[index]
            text = ["<FILL>","<FILL>","<FILL>"] + row.tokenized_text + ["<FILL>","<FILL>","<FILL>"]
            tags = [None,None,None] + row.tags + [None,None,None]

            for _ in range(3):
                idx = random.randrange(3,len(text)-3)

                _tokens = text[idx-3:idx+4]
                _token_tag = tags[idx]

                text_codes = [ translate[token] for token in _tokens ]
                tag_code = tag_translate[_token_tag]

                x.append(text_codes)
                y.append(tag_code)

        x = torch.LongTensor(x)
        y = torch.LongTensor(y)

        optimizer.zero_grad()
        predict = tagger(x)

        loss = torch.nn.functional.cross_entropy(predict,y)
        loss.backward()

        optimizer.step()

        total_loss += loss.item()

    print("epoch loss:",total_loss)
```
