```python
from torch import nn
import torch.nn.functional as F

class AutoEncoder(nn.Module):
    def __init__(self):
        super().__init__()
        self.encoder_a = nn.Linear(50,25)
        self.encoder_b = nn.Linear(25,11)
        self.decoder_a = nn.Linear(11,25)
        self.decoder_b = nn.Linear(25,50)


    def forward(self, x):
        x = self.encoder_a(x)
        x = F.relu(x)
        x = self.encoder_b(x)
        x = F.relu(x)
        x = self.decoder_a(x)
        x = F.tanh(x)
        x = self.decoder_b(x)
        return F.tanh(x)
```

```python
device = (
    "cuda"
    if torch.cuda.is_available()
    else "mps"
    if torch.backends.mps.is_available()
    else "cpu"
)
```

```python
auto_encoder = AutoEncoder().to(device)

optimizer = torch.optim.Adam(auto_encoder.parameters(), lr=0.001)
```

```python
for epoch in range(10):
    total_loss = 0

    for x in dataloader:

        x = x[0]
        y = x

        optimizer.zero_grad()

        predict = auto_encoder(x)

        loss = F.mse_loss(predict,y)
        loss.backward()

        optimizer.step()

        total_loss += loss.item()

    print("epoch loss:",total_loss)
```

```python
def clean_archetype(archetype):
    archetype = str(archetype)

    if archetype == "nan":
        return []

    archetypes = archetype.split("/")

    return [
       x.strip().lower() for x in archetypes
        if "Supports archetypes " not in x
    ]


ds["archetype"] = ds["card_supports"].map(clean_archetype)
ds["archetype"]
```

```python
def clean_anti_archetype(archetype):
    archetype = str(archetype)

    if archetype == "nan":
        return []

    archetypes = archetype.split("/")

    return [
       x.strip().lower() for x in archetypes
        if "Anti-supports archetypes" not in x
    ]


ds["anti_archetype"] = ds["card_anti_supports"].map(clean_anti_archetype)
ds["anti_archetype"]
```

```python
archetypes = functools.reduce(lambda x,y : set(x).union(set(y)),ds["archetype"])
archetypes
```

```python
anti_archetypes = functools.reduce(lambda x,y : set(x).union(set(y)),ds["anti_archetype"])
anti_archetypes
```

```python
intersection = archetypes.intersection(anti_archetypes)
intersection
```

```python
def is_valid(archetype_list):
    for archetype in archetype_list:
        if archetype in intersection:
            return True
    return False

ds["valid_archetype"] = ds.archetype.map(is_valid)
ds["valid_archetype"]
```

```python
ds["valid_anti_archetype"] = ds.anti_archetype.map(is_valid)
ds["valid_anti_archetype"]
```

```python
new_dataset = pd.concat([dataset,archetypes_filters],axis=1)
new_dataset
```

```python
support = {}
antisupport = {}

for _,row in new_dataset.iterrows():
    if not row["valid_archetype"] and not row["valid_anti_archetype"]:
        continue

    X = np.concatenate([row["type"],
                        enc.transform([[row["Attribute"]]]).toarray()[0],
                        np.array([row["link"],
                                  row["rank"],
                                  row["level"],
                                  row["pendulum_scale"],
                                  row["attack"],
                                  row["defense"],
                                  row["tributes"]])],dtype=np.float64)

    for archetype in row["archetype"]:
        support[archetype] = support.get(archetype,[])
        support[archetype].append(X)

    for archetype in row["anti_archetype"]:
        antisupport[archetype] = antisupport.get(archetype,[])
        antisupport[archetype].append(X)

len(support)
```

```python
len(antisupport)
```

```python
import random

anchors_l = []
positives_l = []
negatives_l = []

for anti_archetype in antisupport:
    anti_archetype_cards = antisupport[anti_archetype]

    if not anti_archetype in support:
        continue

    archetype_cards = support[anti_archetype]

    if len(archetype_cards) == 1:
        for negative in anti_archetype_cards:
            i = random.randint(0,len(archetype_cards)-1)
            anchor = archetype_cards[i]
            positive = archetype_cards[i-1]

            anchors_l.append(anchor)
            positives_l.append(positive)
            negatives_l.append(negative)


    if len(anti_archetype_cards) != 1:
        for negative in archetype_cards:
            i = random.randint(0,len(anti_archetype_cards)-1)
            anchor = anti_archetype_cards[i]
            positive = anti_archetype_cards[i-1]

            anchors_l.append(anchor)
            positives_l.append(positive)
            negatives_l.append(negative)
```

```python
class Encoder(nn.Module):
    def __init__(self):
        super().__init__()
        self.encoder_a = nn.Linear(50,25)
        self.encoder_b = nn.Linear(25,11)


    def forward(self, x):
        x = self.encoder_a(x)
        x = F.relu(x)
        x = self.encoder_b(x)
        x = F.relu(x)

        return F.normalize(x,p=2,dim=1)
```

```python
encoder = Encoder().to(device)

optimizer = torch.optim.Adam(encoder.parameters(), lr=0.001)
```

```python
for epoch in range(10):
    total_loss = 0

    for i in range(5,len(anchors_l),5):
        anchors =  torch.Tensor(anchors_l[i-5:i+1])
        positives =  torch.Tensor(positives_l[i-5:i+1])
        negatives =  torch.Tensor(negatives_l[i-5:i+1])

        optimizer.zero_grad()

        anchors =  encoder(anchors)
        positives =  encoder(positives)
        negatives =  encoder(negatives)

        loss = F.triplet_margin_loss(anchors, positives, negatives, margin=0.2)
        loss.backward()

        optimizer.step()

        total_loss += loss.item()

    print("epoch loss:",total_loss)
```
